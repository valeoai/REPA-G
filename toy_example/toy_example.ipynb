## Utils

```python
def sample_from_unnormalized_density(f, N, x_bounds=(-1,1), y_bounds=(-1,1), M=np.e):
    """
    Sample N points from a 2D unnormalized density f(x,y) using rejection sampling.

    Parameters
    ----------
    f : callable
        Function f(x, y) >= 0, bounded by M
    N : int
        Number of samples
    x_bounds : tuple (xmin, xmax)
    y_bounds : tuple (ymin, ymax)
    M : float
        Upper bound on f (default exp(1))

    Returns
    -------
    samples : ndarray of shape (N, 2)
        Accepted samples
    """
    xmin, xmax = x_bounds
    ymin, ymax = y_bounds

    samples = []
    attempts = 0

    while len(samples) < N:
        # 1. sample uniformly on D
        x = np.random.uniform(xmin, xmax)
        y = np.random.uniform(ymin, ymax)

        # 2. sample uniform under the bound
        u = np.random.uniform(0, M)

        # 3. accept / reject
        if u <= f(x, y):
            samples.append([x, y])

        attempts += 1

    samples = np.array(samples)

    acceptance_rate = N / attempts
    print(f"Acceptance rate: {acceptance_rate:.3f}")

    return samples
```

```python
def likelihood(X, c, potential, dataset):
    "compute likelihood of x given condition c"
    i = np.floor(X[:,0] / dataset.cell_size).astype(int)
    j = np.floor(X[:,1] / dataset.cell_size).astype(int)
    p = ((i + j) % 2)   # 1 = active, 0 = empty

    features = dataset.feature_from_coord(X)
    U = potential.compute(torch.from_numpy(features),c.float().cpu()).cpu().detach().numpy()
    pc = np.exp(-U)
    return p*pc
```

```python
import ot

def earth_mover_distance(X, Y):
    """
    Compute Earth Mover Distance (Wasserstein-1) between two 2D point clouds.

    Parameters
    ----------
    X : ndarray of shape (n, d)
    Y : ndarray of shape (m, d)

    Returns
    -------
    emd : float
        Earth Mover Distance
    """
    n = X.shape[0]
    m = Y.shape[0]

    # uniform weights
    a = np.ones(n) / n
    b = np.ones(m) / m

    # cost matrix (Euclidean distance)
    M = ot.dist(X, Y, metric="euclidean")

    # EMD
    emd = ot.emd2(a, b, M)  # squared cost NOT used here (metric already euclidean)

    return emd

def sinkhorn_distance(X, Y, reg=0.05):
    n, m = X.shape[0], Y.shape[0]
    a = np.ones(n) / n
    b = np.ones(m) / m
    M = ot.dist(X, Y)
    return ot.sinkhorn2(a, b, M, reg)
```
